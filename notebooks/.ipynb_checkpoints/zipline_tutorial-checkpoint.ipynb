We could also visualize things by creating an analyze function as seen below.

```python
from zipline.api import order, record, symbol
import matplotlib.pyplot as plt
from matplotlib import style

style.use('ggplot')


def initialize(context):
    pass


def handle_data(context, data):
    order(symbol("AAPL"), 10)
    record(AAPL=data.current(symbol('EBAY'), 'price'))


def analyze(context, perf):
    fig = plt.figure()
    ax1 = fig.add_subplot(111)
    perf.portfolio_value.plot(ax=ax1)
    ax1.set_ylabel('portfolio $ value')
    plt.legend(loc=0)
    plt.show()
```

```python
%zipline --bundle quandl --start 2000-1-1 --end 2005-1-1 -o backtest.pickle --no-benchmark
```

I have personally found that I prefer to keep the backtest and visualization separated, so I can tinker around with the results, rather than needing to re-run everything just to visualize various different things, but, once you have solidified exactly what you want to see, the analyze function could save some time. In other words, handling the visualization outside the pipeline. From this point we should be comfortable using the US-market atleast provided by the quandl dataset.
```python
now = Timestamp.utcnow()
bundle = load('sptest', os.environ, now)
symbols = set(str(asset.symbol) 
              for asset in bundle.asset_finder.retrieve_all(
                           bundle.asset_finder.equities_sids))

print(symbols)
```

```python
!zipline clean -b sptest --keep-last 1
```

```python
def initialize(context):
    set_benchmark(symbol("SPY"))


def handle_data(context, data):
    order(symbol("SPY"), 10)
    record(SPY=data.current(symbol('SPY'), 'price'))
```

Running it for justa year

```python

start_date = pd.Timestamp('2017-01-05', tz='utc')
end_date = pd.Timestamp('2018-03-01', tz='utc')
# performance of algorithm once run
perf = zipline.run_algorithm(start=start_date,
                            end=end_date,
                            trading_calendar=get_calendar('NYSE'),
                            initialize=initialize,
                            capital_base=100000,
                            handle_data=handle_data,
                            bundle='sptest',
                            data_frequency='daily')
```

```python
perf.head()
```

Comparing our strategy to a simple buy and hold strategy. Our strategy is keep buying 10 shares every chance we get. The order_target_percent function is used to allow us to invest a certain percentage of our portfolio value into a company. In this case, the only company we're even consider is the S&P 500, so we're using 1.0 (100%). There are many methods for ordering, this is just one of them. We can do market orders, order specific dollar amounts, order percents, order a target value, and of course cancel pending orders. If we were using the number 1.0 (which is a float) as an argument, then we would be expecting to simply buy/sell 100% of our shares at each step.

```python


style.use("ggplot")

perf.portfolio_value.pct_change().fillna(0).add(1).cumprod().sub(1).plot(label='portfolio')
perf.SPY.pct_change().fillna(0).add(1).cumprod().sub(1).plot(label='benchmark')
plt.legend(loc=2)

plt.show()
```
